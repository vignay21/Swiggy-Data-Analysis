```python
def removing(value):
    if value == '--':
        return np.nan
    else:
        return round(float(value),2)
df['rating']=df['rating'].apply(removing)
```

```python
df['rating'].unique()
```

```python
df['rating'].fillna(round(df['rating'].mean(),2),inplace=True)
```

```python
df.dropna(inplace=True)
df
```

```python
def ends(val):
    val=str(val).split(',')
    
    val=val[-1]
    return str(val)
df['address']=df['address'].apply(ends)
```

```python

def ends1(val):
    val=str(val).split(' ')
    
    val=val[-1]
    return str(val).lower()
df['address']=df['address'].apply(ends1)
df
```
